```python
import pandas as pd
import scrapper
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('all_shows_data.csv')
df.head()
```

```python
def get_page_number(url):
        page = requests.get(url)

        soup = BeautifulSoup(page.content, "html.parser")
        results = soup.find_all(class_="page-item last")
       
        if not results:
             return 1
        
        for result in results:
            link = result.find('a')['href']
            return int(link.split('=')[-1])
            
        
        
```

```python
user_list=[]
for link in df.Link:
    url = "https://mydramalist.com/%s/reviews?page=1" % (link)
    page_num = get_page_number(url)
    
    for i in range(1,page_num+1):
        url = "https://mydramalist.com/%s/reviews?page=%s" % (link,i) 
        page = requests.get(url)

        soup = BeautifulSoup(page.content, "html.parser")
        results = soup.find_all(class_="review")
        

        for result in results:
            name = result.find('a')['href']
            user_list.append(name.split('/')[-1])
            


        
```

```python
print(user_list)
```

```python
import scrapper
ratings_df = pd.DataFrame()
scr = scrapper.Scrapper()


for user in user_list:
    try:
        temp_df = scr.get_user_list(user) 
    except AttributeError :
        print("nie ma" + user)
    else:   
        temp_df['User'] = user
        ratings_df = pd.concat([ratings_df,temp_df],axis=0)



```
